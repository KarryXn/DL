```python
import matplotlib.pyplot as plt
import pandas as pd

# 假设 CSV 文件路径为 'log_data.csv'
csv_file_path = r'checkpoints\Meanteacher_unet\training_summary.csv'

# 从CSV文件读取数据
df = pd.read_csv(csv_file_path)

# 检查数据是否正确读取
print(df.head())

# 绘制 train dice loss 和 eval dsc 的折线图
plt.figure(figsize=(12, 6))

# 绘制train dice loss
plt.plot(df['step'], df['train loss'], label='Train Dice Loss', color='blue', marker='o')

# # 绘制eval dsc
# plt.plot(df['step'], df['eval dsc'], label='Eval DSC', color='red', marker='o')

# 设置标题和标签
plt.title('UNet')
plt.xlabel('Steps')
plt.ylabel('Value')
plt.xticks(rotation=45)

# 添加图例
plt.legend()

# 显示图像
plt.tight_layout()
plt.show()
#Train dice loss: 0.026139235430293613  Eval DSC: 0.21158538927818038  Train time: 27.09419012069702
```
