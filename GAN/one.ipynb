```python
import torch

# 检查 PyTorch 是否安装成功
print("PyTorch version:", torch.__version__)

# 检查 CUDA 是否可用
print("CUDA available:", torch.cuda.is_available())

# 如果有多块GPU，可查看CUDA设备数
if torch.cuda.is_available():
    print("CUDA device count:", torch.cuda.device_count())
    print("Current CUDA device:", torch.cuda.current_device())
    print("CUDA device name:", torch.cuda.get_device_name(0))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 CSV 文件
df = pd.read_csv('./training_loss.csv')

# 检查数据结构
print(df.head())  # 查看数据前几行以确保格式正确

# 假设 CSV 文件包含列名 'epoch', 'd_loss', 'g_loss', 'D(x)', 'D(G(z))'
# 用相应的列名绘制折线图
plt.figure(figsize=(12, 8))

# 绘制折线图
plt.plot(df['epoch'], df['d_loss'], label='d_loss', color='blue', marker='o')
plt.plot(df['epoch'], df['g_loss'], label='g_loss', color='green', marker='o')
plt.plot(df['epoch'], df['D(x)'], label='D(x)', color='red', marker='o')
plt.plot(df['epoch'], df['D(G(z))'], label='D(G(z))', color='purple', marker='o')

# 添加标题和轴标签
plt.title('GAN Training Progress')
plt.xlabel('Epoch')
plt.ylabel('Loss / Probability')
plt.legend()
plt.grid()

# 显示图形
plt.show()
```
